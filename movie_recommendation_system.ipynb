## Mainly three types of recommendation systems in machine learning based on filtering are used to suggest product and services to the consumers.

### 1. Content Filtering

### 2. Collaborative Filtering

### 3. Hybrid Filtering

## 1. Content Filtering:

   In this algorithm, we try finding items look alike. Once we have item look like matrix,

      we can easily recommend alike items to a customer, who has purchased any item from the store.

## 2. Collaborative Filtering:

      Here, we try to search for look alike customers and offer products based on what his/her lookalike has chosen.

      This algorithm is very effective but takes a lot of time and resources.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.neighbors import NearestNeighbors
from scipy.sparse import csr_matrix

# Загрузим данные
tag = pd.read_csv('/content/tag.csv')
rating = pd.read_csv('/content/rating.csv')
movies = pd.read_csv('/content/movie.csv')

# Считаем количество оценок для каждого пользователя
user_rating_count = rating['userId'].value_counts()

# Фильтруем только самых активных пользователей (например, 500 и более оценок)
active_users = user_rating_count[user_rating_count >= 500].index

# Оставляем только данные для этих пользователей
rating = rating[rating['userId'].isin(active_users)]

# Объединяем данные
movie_details = movies.merge(rating, on='movieId')
movie_details.drop(columns=['timestamp'], inplace=True)

# Вычисляем количество оценок для каждого фильма
number_rating = movie_details.groupby('title')['rating'].count().reset_index()
number_rating.rename(columns={'rating': 'number of rating'}, inplace=True)

# Объединяем с количеством оценок
df = movie_details.merge(number_rating, on='title')

# Фильтруем фильмы, у которых больше 50 оценок
df = df[df['number of rating'] >= 50]
df.drop_duplicates(['title', 'userId'], inplace=True)

# Создаем сводную таблицу (пользователи x фильмы)
df['rating'] = df['rating'].astype(int)
movie_pivot = df.pivot_table(columns='userId', index='title', values='rating')

# Заполняем пропуски нулями
movie_pivot.fillna(0, inplace=True)

# Преобразуем в разреженную матрицу
movie_sparse = csr_matrix(movie_pivot)

# Инициализируем модель ближайших соседей
model = NearestNeighbors(n_neighbors=7, algorithm='brute', metric='cosine')
model.fit(movie_sparse)

# Реализуем k-fold кросс-валидацию
def leave_n_out_cross_validation(df, n=5):
    all_rmse = []
    all_mae = []

    # Разбиваем данные на n фолдов
    unique_movies = df.index
    movie_fold_size = len(unique_movies) // n

    for i in range(n):
        # Разбиваем на тренировочные и тестовые данные
        test_movies = unique_movies[i * movie_fold_size: (i + 1) * movie_fold_size]
        train_movies = unique_movies[~np.isin(unique_movies, test_movies)]

        # Создаем тренировочные и тестовые данные
        train_data = df.loc[train_movies]
        test_data = df.loc[test_movies]

        # Обучаем модель на тренировочных данных
        model.fit(csr_matrix(train_data))

        # Делаем предсказания на тестовых данных
        distances, suggestions = model.kneighbors(csr_matrix(test_data))

        # Теперь для каждого тестового фильма будем собирать предсказанные рейтинги
        predicted_ratings = []
        true_ratings = []

        for idx, movie in enumerate(test_movies):
            true_ratings.extend(test_data.loc[movie].values)

            # Получаем рейтинги ближайших соседей для текущего фильма
            neighbors = suggestions[idx]
            neighbor_ratings = train_data.iloc[neighbors, :].values

            # Для каждого пользователя в тестовой выборке, берем среднее из соседей
            avg_neighbor_ratings = np.mean(neighbor_ratings, axis=0)
            predicted_ratings.extend(avg_neighbor_ratings)

        # Вычисляем RMSE/MAE для этой итерации
        rmse = np.sqrt(mean_squared_error(true_ratings, predicted_ratings))
        all_rmse.append(rmse)

        mae = mean_absolute_error(true_ratings, predicted_ratings)
        all_mae.append(mae)

    avg_rmse = np.mean(all_rmse)
    avg_mae = np.mean(all_mae)

    print(f'Average RMSE: {avg_rmse}')
    print(f'Average MAE: {avg_mae}')

# Запускаем кросс-валидацию
leave_n_out_cross_validation(movie_pivot, n=5)
```

```python
def get_similar_movies(movie_name, n_neighbors=7):
    """
    Функция для вывода похожих фильмов на основе названия фильма.

    Параметры:
    - movie_name (str): Название фильма, для которого нужно найти похожие фильмы.
    - n_neighbors (int): Количество похожих фильмов для вывода. По умолчанию 7.

    Возвращает:
    - Список похожих фильмов.
    """
    # Найдем ID фильма по названию
    #movie_id = movie_pivot.index.get_loc(movie_name)
    movie_id = df[df['title'] == movie_name].drop_duplicates('title')['movieId'].values[0]
    print(movie_id)
    # Получаем индекс строки в сводной таблице movie_pivot
    movie_index = movie_pivot.index.get_loc(movie_name)

    # Находим ближайших соседей для этого фильма
    distances, suggestions = model.kneighbors(movie_pivot.iloc[movie_index, :].values.reshape(1, -1), n_neighbors=n_neighbors)

    # Выводим похожие фильмы
    similar_movies = []
    for i in range(len(suggestions[0])):
        similar_movie = movie_pivot.index[suggestions[0][i]]
        similar_movies.append(similar_movie)
    if movie_name in similar_movies:
        similar_movies.remove(movie_name)
    return similar_movies

# Пример использования функции:
movie_name = "Insidious (2010)"  # Укажите название фильма
similar_movies = get_similar_movies(movie_name)

print(f"Похожие фильмы для '{movie_name}':")

for movie in similar_movies:
    print(movie)
```

```python
def get_similar_movies_by_id(movie_id, n_neighbors=7):
    """
    Функция для вывода похожих фильмов на основе ID фильма.

    Параметры:
    - movie_id (int): ID фильма, для которого нужно найти похожие фильмы.
    - n_neighbors (int): Количество похожих фильмов для вывода. По умолчанию 7.

    Возвращает:
    - Словарь, где ключ — это ID похожего фильма, а значение — расстояние (distance) до него.
    """
    # Проверяем, присутствует ли фильм с данным ID в сводной таблице
    if movie_id not in movie_pivot.index:
        print(f"Фильм с ID {movie_id} не найден в pivot-таблице.")
        return {}

    # Получаем индекс строки в pivot-таблице
    movie_index = movie_pivot.index.get_loc(movie_id)

    # Находим ближайших соседей для этого фильма
    distances, suggestions = model.kneighbors(
        movie_pivot.iloc[movie_index, :].values.reshape(1, -1),
        n_neighbors=n_neighbors
    )

    # Составляем словарь похожих фильмов
    similar_movies = {
        movie_pivot.index[suggestions[0][i]]: distances[0][i]
        for i in range(len(suggestions[0]))
        if movie_pivot.index[suggestions[0][i]] != movie_id  # Исключаем сам фильм
    }

    return similar_movies

# Пример использования функции:
movie_id = 85788  # Укажите ID фильма (например, "Toy Story" имеет ID 1)
similar_movies = get_similar_movies_by_id(movie_id, 2)

if similar_movies:
    print(f"Похожие фильмы для ID {movie_id}:")
    for similar_movie_id, distance in similar_movies.items():
        print(f"ID: {similar_movie_id}, Distance: {distance:.4f}")
```

```python
def get_similar_movies_by_title(movie_title, n_neighbors):
    """
    Функция для вывода похожих фильмов на основе названия фильма (title).

    Параметры:
    - movie_title (str): Название фильма, для которого нужно найти похожие фильмы.
    - n_neighbors (int): Количество похожих фильмов для вывода. По умолчанию 7.

    Возвращает:
    - Словарь, где ключ — это название похожего фильма, а значение — расстояние (distance) до него.
    """
    # Проверяем, присутствует ли фильм с данным названием в таблице
    if movie_title not in movie_pivot.index:
        print(f"Фильм '{movie_title}' не найден в pivot-таблице.")
        return {}

    # Получаем индекс строки в pivot-таблице
    movie_index = movie_pivot.index.get_loc(movie_title)

    # Находим ближайших соседей для этого фильма
    distances, suggestions = model.kneighbors(
        movie_pivot.iloc[movie_index, :].values.reshape(1, -1),
        n_neighbors=n_neighbors
    )

    # Составляем словарь похожих фильмов
    similar_movies = {
        movie_pivot.index[suggestions[0][i]]: distances[0][i]
        for i in range(len(suggestions[0]))
        if movie_pivot.index[suggestions[0][i]] != movie_title  # Исключаем сам фильм
    }

    return similar_movies

# Пример использования функции:
movie_title = "Insidious (2010)"  # Укажите название фильма
similar_movies = get_similar_movies_by_title(movie_title, 10)

if similar_movies:
    print(f"Похожие фильмы для '{movie_title}':")
    for similar_movie_title, distance in similar_movies.items():
        print(f"Название: {similar_movie_title}, Distance: {distance:.4f}")
```

```python
def predict(user_id, n_neighbors):
    """
    Функция предсказания фильмов для пользователя на основе его оценок и похожести фильмов.

    Параметры:
    - user_id (int): ID пользователя, для которого нужно предсказать фильмы.
    - n_neighbors (int): Количество похожих фильмов для каждого фильма. По умолчанию 7.

    Возвращает:
    - Список фильмов, отсортированный по расстоянию (distance) в порядке убывания.
    """
    # Проверяем, есть ли пользователь в данных
    if user_id not in movie_pivot.columns:
        print(f"Пользователь с ID {user_id} не найден.")
        return []

    # Получаем фильмы и оценки, связанные с пользователем
    user_ratings = movie_pivot[user_id]

    # Словарь для хранения скорингов
    res = {}

    # Проходим по фильмам, которые пользователь оценил
    for movie_title, score in user_ratings.items():
        if score < 4:  # Игнорируем фильмы, которые пользователь не оценил
            continue

        # Получаем похожие фильмы для текущего фильма
        similar_movies = get_similar_movies_by_title(movie_title, n_neighbors)

        # Добавляем взвешенные расстояния для похожих фильмов
        for similar_movie_title, distance in similar_movies.items():
            weighted_distance = distance * score  # Учитываем оценку пользователя
            if similar_movie_title not in res:
                res[similar_movie_title] = weighted_distance
            else:
                res[similar_movie_title] += weighted_distance  # Суммируем веса для фильмов

    # Сортируем результат по убыванию distance
    sorted_res = sorted(res.items(), key=lambda x: x[1], reverse=True)

    # Возвращаем отсортированный список
    return sorted_res

# Пример использования:
user_id = 11  # Укажите ID пользователя
recommendations = predict(user_id, 2)

if recommendations:
    print("Рекомендации:")
    for movie_title, weighted_distance in recommendations:
        print(f"Название: {movie_title}, Weighted Distance: {weighted_distance:.4f}")
```
